#### Writing comments in Python

#### Does Python know about your error before it runs your code?


Python is what is called an <em>interpreted language</em>. Compiled languages examine your entire program at compile time, and are able to warn you about a whole class of errors prior to execution. In contrast, Python interprets your script line by line as it executes it. Python will stop executing the entire program when it encounters an error (unless the error is expected and handled by the programmer, a more advanced subject that we'll cover later on in this course).

### Python Variable Naming Conventions
In Python, variable names play a crucial role in code readability and maintainability. Following proper naming conventions not only makes your code more understandable but also helps in avoiding errors and confusion. Here are some guidelines for naming variables in Python:

1. **The first character cannot be a digit**.

2. **Use descriptive names:** Choose names that clearly describe the purpose or content of the variable. This makes it easier for others (and your future self) to understand the code.

3. **Use lowercase with underscores:** For most variables, use lowercase letters and separate words with underscores (`snake_case`). This convention enhances readability, especially for longer variable names.

4. **Avoid single-letter names:** Unless used in specific contexts like loop counters (`i`, `j`, `k`), avoid single-letter variable names as they are often unclear and can lead to confusion.

5. **Be consistent:** Maintain consistency in naming across your codebase. If you're working on a project with existing naming conventions, adhere to them for uniformity.

6. **Avoid reserved keywords:** Don't use Python keywords (e.g., `if`, `else`, `for`, `while`, `import`, etc.) as variable names to prevent syntax errors.

````python
# String
name = "Soliman"
multi_line_string_as_comment = """This is a
multi-line string."""

print(name)
print(multi_line_string_as_comment)
````

````python

# Basic String Operations

# Concatenation
str1 = "Hello"
str2 = "World"

concatenated = str1 + " " + str2

print("Concatenation: ", concatenated)
````

````python
# Repetition
repeated = str1 * 3
print("Repetition: ", repeated)
````

````python

# Slicing
sliced = concatenated[0:5]
print("Slicing: ", sliced)
````

````python

# Length
length = len(concatenated)
print("Length: ", length)
````

````python

# Membership
is_in = "World" in concatenated
print("Membership (in): ", is_in)
````

````python

# String Methods

# Upper case
upper_str = concatenated.upper()
print("Upper case: ", upper_str)
````

````python

# Lower case
lower_str = concatenated.lower()
print("Lower case: ", lower_str)
````

````python

# Capitalize
capitalized_str = concatenated.capitalize()
print("Capitalize: ", capitalized_str)
````

````python

# Title case
title_str = concatenated.title()
print("Title case: ", title_str)
````

````python

# Strip
strip_str = "   Hello World   ".strip()
print("Strip: ", strip_str)
````

````python

# Replace
replaced_str = concatenated.replace("World", "Python")
print("Replace: ", replaced_str)
````

````python

# Split
split_str = concatenated.split(" ")
print("Split: ", split_str)
````

````python

# Join
joined_str = "-".join(split_str)
print("Join: ", joined_str)
````

````python

# Find
found_index = concatenated.find("World")
print("Find: ", found_index)
````

````python

# Format
formatted_str = "Hello, {}!".format("Python")
print("Format: ", formatted_str)
````

````python

# Format _ 2nd method
var= 'partner'
formatted_str2 = f"Hello, {var}!"
print("Format: ", formatted_str2)
````

### Arithmetic operations

````python

# Basic Operations

# Arithmetic operations
x = 10
y = 5
print("Addition:", x + y)
print("Subtraction:", x - y)
print("Multiplication:", x * y)
print("Division:", x / y)
print("Floor Division:", x // y)
print("Modulus:", x % y)
print("Exponentiation:", x ** y)
````

### Comparison operations

````python

# Comparison operations
print("Equal:", x == y)
print("Not equal:", x != y)
print("Greater than:", x > y)
print("Less than:", x < y)
print("Greater than or equal to:", x >= y)
print("Less than or equal to:", x <= y)
````

### Logical operations

````python

# Logical operations
print("And:", x > 5 and y < 10)
print("Or:", x > 5 or y > 10)
print("Not:", not (x > 5))
````

### Bitwise operations

````python

# Bitwise operations

print(bin(x)[2:],bin(y)[2:],'\n')

print("Bitwise AND:", x & y)
print("Bitwise OR:", x | y)
````

### Assignment operations

````python

# Assignment operations
z = x + y
z += x
print("Assignment (z += x):", z)
z -= y
print("Assignment (z -= y):", z)
z *= x
print("Assignment (z *= x):", z)
z /= y
print("Assignment (z /= y):", z)
z %= x
print("Assignment (z %= x):", z)
z **= y
print("Assignment (z **= y):", z)
````
